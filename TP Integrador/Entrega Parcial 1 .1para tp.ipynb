```
import matplotlib.pyplot as plt
```

```
sns.pairplot(ip)#, hue='evento_nombre')
```

```
plt.figure(figsize=(12, 6))
sns.scatterplot(data=ip, 
                x='semanas_epidemiologicas', 
                y='cantidad_casos', 
                hue='evento_nombre', 
                size='cantidad_casos',
                sizes=(20, 500), 
                alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Evento')
plt.title('Dispersión de Casos por Semana epidemiologica y Tipo de Evento')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()

plt.show()
```

```
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8)) 

tabla_frecuencia = pd.crosstab(ip['grupo_edad_desc'], ip['evento_nombre'])

sb.heatmap(tabla_frecuencia, annot=True, fmt='d', cmap='YlGnBu')

plt.title('Densidad de Casos: Evento vs Grupo de Edad')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Grupo de Edad')
plt.xlabel('Evento')
plt.tight_layout()

plt.show()
```
