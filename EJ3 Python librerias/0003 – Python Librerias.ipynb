## Ejercicio 2 – Dataset simple con Pandas

```
cuadro ={
    "nombre": ["Ana","Juan","Carla","Pedro","Lucia"],
    "notas": [8,6,9,5,7]
}
df = pd.DataFrame(cuadro)
```

```
print(df)
```

```
promedio=df['notas'].sum()/df['notas'].count()
maxima=df['notas'].max()
minima=df['notas'].min()
```

```
print("El promedio es: ", promedio)
print("El máximo es: ", maxima)
print("El mínimo es: ", minima)
```
