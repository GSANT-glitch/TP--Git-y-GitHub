## Ejercicio 1 – Variables y operaciones

## Ejercicio 2 – Condicionales

```
edad=int(input("Por favor ingrese su edad: "))

if edad >=18:
    print("Su edad es",edad,"años, usted es mayor de edad")
else:
    print("Su edad es",edad,"años, usted es menor de edad")
```

## Ejercicio 3 – Listas y bucles

```
lista1=[10,9,8,7,6,5]
```

```
suma=0
for i in lista1:
    suma += i
prom=suma/len(lista1)
```

```
print("La suma de la lista es:", suma)
print("El promedio de la lista es:", prom)
```
